```python
import requests
import pandas as pd

# Your EODHD API token
api_token = '6847ec4b1ce359.56687473'

# API endpoint for historical data
url = 'https://eodhd.com/api/eod/AAPL.US'

# Parameters for the API call
params = {
    'api_token': api_token,
    'from': '2024-03-01',
    'to': '2025-02-28',
    'period': 'd',        # daily data
    'fmt': 'json'         # response format
}

# Make the request
response = requests.get(url, params=params)

# Convert to JSON
data = response.json()

# Load data into DataFrame
df = pd.DataFrame(data)

# Optional: Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Sort by date (optional)
df = df.sort_values('date')

# Display the first few rows
print(df.head())

# Save to CSV (optional)
df.to_csv('AAPL_Historical_Mar2024_to_Feb2025.csv', index=False)
```
